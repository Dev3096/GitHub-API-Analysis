<h3> Function to get repo Issue data <h3>

```python
import datetime
def getRepoIssueData(url):
    ACCESS_TOKEN = '<your access token>'
    headers = {
    'Authorization': f'token {ACCESS_TOKEN}',
    'Accept': 'application/vnd.github.v3+json'
    }
    response = requests.get(url, headers=headers)
    issues = response.json()
    return issues
```

<h3> Function to get Dates for March, 2022 <h3>

```python
import datetime
def getMarch(issues):
    march_date = []
    for issue in issues:
        created_at = datetime.datetime.strptime(issue, '%Y-%m-%dT%H:%M:%SZ')
        month = created_at.strftime("%B")
        if (month == "March"):
            march_date.append(issue)
    return march_date
```

<h3> Function to get total number of issues created for every Monday for the month of March, 2022 <h3> 

```python
import datetime
def getMonday(issues):
    issues_per_monday = {}
    for issue in issues:
        created_at_dt = datetime.datetime.strptime(issue, "%Y-%m-%dT%H:%M:%SZ")
        day_of_week = created_at_dt.strftime("%A")
        if day_of_week == "Monday":
        # Get the week number of the creation date
            week_number = created_at_dt.strftime("%U")
        # Check if the week number is already in the issues_per_monday dictionary
            if week_number in issues_per_monday:
            # Increment the count for the week number in the issues_per_monday dictionary
                issues_per_monday[week_number] += 1
            else:
                # Set the count for the week number to 1 in the issues_per_monday dictionary
                issues_per_monday[week_number] = 1
    return issues_per_monday
```

<h3> Function to get total number of forks and pulls created in 2022 <h3> 

```python
import datetime
def dataPullsForksCreated(forks):
    forks_2022 = []
    for fork in forks:
        created_at = datetime.datetime.strptime(fork['created_at'], '%Y-%m-%dT%H:%M:%SZ')
        if (created_at.year == 2022):
            forks_2022.append(fork['created_at'])
    return forks_2022
```

<h3> Function to get total number of issues created every month for year 2022 <h3> 

```python
import datetime
def issuesCrestedEveryMonthNew(issues):
# Dictionary to hold months and count of issues created in that month
    per_month = {}
#     for i in range (1,13):
#         myDict[i] = 0
    # print(myDict)
    for issue in issues:
        created_at = datetime.datetime.strptime(issue, '%Y-%m-%dT%H:%M:%SZ')
        month  = created_at.strftime("%B")
        if month in per_month:
            per_month[month] += 1
        else:
            per_month[month] = 1
    return per_month
```

<h3> Functon to get total number of issues created every DAY of the Week (total on Mondays, Tuesdays, etc.) for year 2022 <h3>

```python
import datetime
def issuesPerWeek(issues):
    issue_counts = {'Monday': 0, 'Tuesday': 0, 'Wednesday': 0, 'Thursday': 0, 'Friday': 0, 'Saturday': 0, 'Sunday': 0}
    for issue in issues:
        created_at = datetime.datetime.strptime(issue, '%Y-%m-%dT%H:%M:%SZ')
        day_of_week = created_at.strftime('%A')
        if (day_of_week in issue_counts):
            issue_counts[day_of_week] +=1
    return issue_counts
```
